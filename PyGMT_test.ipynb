```python
import pygmt
from pygmt.datasets import load_earth_relief

# Define the region for Taiwan
taiwan_region = "119/123/21/26"

# Load earth relief data for Taiwan. We specify the region for efficient loading.
# '15s' (15 arc-seconds) is a very high resolution for relief data.
grid = load_earth_relief(resolution='15s', registration='pixel', region=taiwan_region)

fig = pygmt.Figure()

# Set up the basemap for Taiwan
fig.basemap(
    region=taiwan_region,
    projection='M15c', # Mercator projection, 15 cm wide, suitable for regional maps
    frame="ag" # Add annotations and gridlines
)

# Plot the relief image
fig.grdimage(grid, cmap='geo', shading=True)

# Add coastlines and borders
fig.coast(
    resolution='f', # Changed resolution to 'f' for full resolution coastlines (highest available)
    shorelines="1/0.5p", # Show shorelines with a thin pen
    # Removed 'land' and 'water' coloring to allow grdimage colors to show
    borders="1/0.5p", # Show national borders
    area_thresh=1000 # Minimum area for features to be plotted
)

# Add a colorbar for the relief, with a label
fig.colorbar(frame='a+l"Elevation"')

fig.show()
```

```python
!pip install pygmt
import pygmt
import pandas as pd
import numpy as np
import os

# ÂàùÂßãÂåñÁπ™Âúñ
fig = pygmt.Figure()

# Ë®≠ÁΩÆÁπ™ÂúñÂçÄÂüüÂíåÂèÉÊï∏
region = "-100/-60/-50/10"
projection = "M15c"
frame = ["a", "+tNazca Plate Tectonics and Seismicity"]

# ÊùøÂ°äÈÇäÁïåÊï∏ÊìöÊ∫êË∑ØÂæë (‰ΩøÁî®ËÄÖÊèê‰æõ)
plate_data_file = "/content/PB2002_boundaries.dig.txt"

# --- Ê≠•È©ü 0ÔºöÊï∏ÊìöÈ†êËôïÁêÜ (‰øÆÊ≠£ AWK ÁØ©ÈÅ∏Ë™ûÊ≥ï) ---
print("Starting data pre-processing (gmt awk)...")

# ÂÅáË®≠ÈÇäÁïåÈ°ûÂûã‰ª£Á¢º (1, 2, 3) Á∑äË∑üÂú®Ë°åÈ¶ñÁöÑÂ§öÊÆµÁ∑öÊ®ôÈ†≠ (> Êàñ > #) Âæå
# 1. ÁØ©ÈÅ∏ËÅöÂêàÂûãÈÇäÁïå (Type 1)
# ÂåπÈÖçÊâÄÊúâ‰ª• > ÈñãÈ†≠ÔºåÈö®ÂæåË∑üËëó‰ªªÊÑèÂ≠óÁ¨¶ÂíåÊï∏Â≠ó 1 ÁöÑË°å
!gmt awk "/^>.*1/" {plate_data_file} > pb2002_type1.gmt

# 2. ÁØ©ÈÅ∏‰º∏ÂºµÂûãÈÇäÁïå (Type 2)
!gmt awk "/^>.*2/" {plate_data_file} > pb2002_type2.gmt

# 3. ÁØ©ÈÅ∏ËΩâÊèõÂûãÈÇäÁïå (Type 3)
!gmt awk "/^>.*3/" {plate_data_file} > pb2002_type3.gmt

print("Data pre-processing complete.")
# ----------------------------------------

# Ê≠•È©ü A, B, C: Áπ™Ë£ΩÂú∞ÂΩ¢„ÄÅÊµ∑Â≤∏Á∑ö„ÄÅÊùøÂ°äÈÇäÁïå
fig.grdimage(
    grid="@earth_relief_15m",
    region=region,
    projection=projection,
    shading=True,
    cmap="oleron",
    frame=frame
)


fig.coast(
    resolution="h",
    shorelines=True,
    borders="1/0.5p,black",
    area_thresh="500",
    land="#F5F5DC"
)

# C.1: ‰øØË°ùÂ∏∂ (ËÅöÂêàÂûãÈÇäÁïå, Type=1)
fig.plot(
    data="pb2002_type1.gmt",
    pen="1p,red",
    style="f0.3c/0.1c+l+t", # ***‰øÆÊ≠£Èªû***: ‰øùÁïô style Áπ™Ë£Ω‰∏âËßíÂΩ¢
    fill="red3",
    no_clip=True
)

# C.2: ‰º∏ÂºµÂûãÈÇäÁïå (ÂàÜÈõ¢ÂûãÈÇäÁïå, Type=2)
fig.plot(
    data="pb2002_type2.gmt",
    pen="1p,blue,dashed"
)

# C.3: ËΩâÊèõÂûãÈÇäÁïå (ËΩâÊèõÊñ∑Â±§, Type=3)
fig.plot(
    data="pb2002_type3.gmt",
    pen="1p,black,dotted"
)

# Ê≠•È©ü D, E: Áπ™Ë£ΩÁÅ´Â±±ÂíåÂú∞ÈúáÊï∏Êìö
volcanoes = {
    'lon': [-68.54, -78.43, -70.92, -73.57],
    'lat': [-27.11, -0.68, -38.69, -16.29],
}
fig.plot(x=volcanoes['lon'], y=volcanoes['lat'], style="kstar/0.2c", fill="orange", pen="0.5p,black")

historical_quakes = pd.DataFrame({
    'lon': [-73.05, -71.60, -78.73, -70.36],
    'lat': [-38.2, -35.8, -17.8, -19.6],
    'mag': [9.5, 8.8, 9.0, 8.8],
    'year': [1960, 2010, 1868, 1877]
})

fig.plot(
    x=historical_quakes['lon'],
    y=historical_quakes['lat'],
    size=historical_quakes['mag'] * 0.05,
    style="c",
    fill="blue@50",
    pen="0.5p,blue",
    no_clip=True
)

# Ê≠•È©ü F: Ê∑ªÂä†Ëâ≤Ê®ôËàáÂúñ‰æã

fig.colorbar(frame=["a5000", "x+lElevation (m)"])

# ***ÊúÄÁµÇÂúñ‰æã‰øÆÊ≠£***Ôºö‰øÆÊ≠£Âúñ‰æãÂÆöÁæ©Êñá‰ª∂‰∏≠ÁöÑË™ûÊ≥ïÔºåÁßªÈô§‰∏çÈúÄË¶ÅÁöÑÂñÆÂÖÉÔºå‰ª•ÈÅøÂÖçË≠¶Âëä„ÄÇ
legend_text = f"""G 0.2c
S 0.5c - 1p,red 0.5c f0.3c/0.1c+l+t red3 Subduction Zone (Convergent)
S 0.5c - 1p,blue,dashed 0.5c - Spreading Ridge (Divergent)
S 0.5c - 1p,black,dotted 0.5c - Transform Fault (Strike-slip)
G 0.2c
S 0.5c kstar/0.2c orange 0.5p,black Volcanoes (Andean Arc)
S 0.5c c/{(9.5 * 0.05):.2f}c blue@50 0.5p,blue Historical M>8 Quakes (Size $\propto$ Mag)
"""

# ÂØ´ÂÖ•Âúñ‰æãÂÆöÁæ©Êñá‰ª∂
legend_filename = "legend_spec.txt"
with open(legend_filename, "w") as f:
    f.write(legend_text)

# ÂÉÖË™øÁî®‰∏ÄÊ¨° fig.legend
fig.legend(
    position="JBL+o0.2c/0.2c+w4c/4.5c",
    box="+gwhite+p1p",
    spec=legend_filename
)


# Ê≠•È©ü G: È°ØÁ§∫Áπ™Âúñ
fig.show()
```
